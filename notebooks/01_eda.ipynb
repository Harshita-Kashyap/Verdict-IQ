Data Cleaning 

```python
train_df = train_df.drop(columns=['Unnamed: 0'])
test_df = test_df.drop(columns=['Unnamed: 0'])
```

```python
train_df['reviewContent'] = train_df['reviewContent'].fillna('')
test_df['reviewContent'] = test_df['reviewContent'].fillna('')

train_df['location'] = train_df['location'].fillna('Unknown')
test_df['location'] = test_df['location'].fillna('Unknown')
```

```python
train_df['yelpJoinDate'].head(10)
```

```python
train_df['date'] = pd.to_datetime(
    train_df['date'],
    format='%m/%d/%Y'
)

train_df['yelpJoinDate'] = pd.to_datetime(
    train_df['yelpJoinDate'],
    format='%Y-%m-%d'
)

test_df['date'] = pd.to_datetime(
    test_df['date'],
    format='%m/%d/%Y'
)

test_df['yelpJoinDate'] = pd.to_datetime(
    test_df['yelpJoinDate'],
    format='%Y-%m-%d'
)
```

```python
train_df['reviewContent'] = (
    train_df['reviewContent']
    .str.lower()
    .str.strip()
)

test_df['reviewContent'] = (
    test_df['reviewContent']
    .str.lower()
    .str.strip()
)
```

```python
train_df.to_csv(
    '../data/cleaned/train_cleaned.csv',
    index=False
)

test_df.to_csv(
    '../data/cleaned/test_cleaned.csv',
    index=False
)

print("Cleaned datasets saved successfully.")
```
